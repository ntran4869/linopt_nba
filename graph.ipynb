```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('pts.csv')
data.head()
```

```python
# Adding the current year to the Date column and converting to datetime
current_year = datetime.now().year
data['Date'] = pd.to_datetime(data['Date'] + f"/{current_year}", format='%m/%d/%Y')

# Reattempting to plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Points Got'], label='Points Got', marker='o')
plt.plot(data['Date'], data['Expected Get'], label='Expected Get', marker='o')
plt.plot(data['Date'], data['min Pts to win $'], label='Min Pts to Win', marker='o')

# Adding labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Points')
plt.title('Points Comparison Over Time')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Rotating date labels for better visibility
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['diff'], label='Range To Winning', marker='o', color='red')

plt.axhline(0, color='green', linestyle='--', linewidth=1, label='Zero Line')

# Adding labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Points')
plt.title('Points Difference with Winning')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Rotating date labels for better visibility
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
